```python
class InvalidSalException(BaseException):
    def __init__(self,msg):
        self.message = msg

class Emp:
    __ecount = 0   #variable  start with __  private static
    
    def __str__(self):
        return (" STR: Name: {} No: {} Sal: {} City: {}".format(self.EmpName,self.EmpNo,self.EmpSal,self.EmpCity))
    
    def __repr__(self):
        return "REPR: Name: {} No: {} Sal: {} City: {}".format(self.EmpName,self.EmpNo,self.EmpSal,self.EmpCity)
        
    def __del__(self):
        print("Emp",self.EmpName, " is being destroyed")
        
    def __init__(self,enm=None,esal=10000,ecity="Pune"):  # self ? = in python self means this = pointer to the  current instance
        self.EmpName = enm
        self.EmpSal = esal
        self.EmpDa = self.EmpSal * 0.25
        self.EmpCity = ecity
        Emp.__ecount = Emp.__ecount + 1
        self.EmpNo = Emp.__ecount
        self.__SecretCode = self.EmpName + ":" + self.EmpName 
        
    def __gt__(self,other):
        return self.EmpSal > other.EmpSal
        
    #we do not want salary in negative and we want sal value min 10000,
    # we can solve this problem by getter setter method
    @property
    def EmpSal(self):#getter
        return self.__empsal
           
    @EmpSal.setter
    def EmpSal(self,value):
        if value >= 10000:
            self.__empsal = value
        else:
            raise InvalidSalException("Sal should be min 10000")
    
    def __AfterDisplay(self):
        print("********* Display Done *********")
        
    def Display(self):
        print("EmpNo:",self.EmpNo, "Name:",self.EmpName, "Sal:",self.EmpSal, "DA:",self.EmpDa)
        
        self.__AfterDisplay()
        
    @staticmethod
    def GetEmpCount():
        return Emp.__ecount
    
class OnsiteEmp(Emp): #inheriting Emp class
    def __init__(self,enm=None,esal=1000,ecity="mumbai",visano="not found"):
        Emp.__init__(self,enm,esal,ecity)
        self.visa=visano
    def Display(self):
        super().Display()
        print("VisaNo of the emp: {}".format(self.visa))

def empsortattr(e):
    return e.EmpName
```

```python
        
def main():
    e1 = Emp ("Abhishek", 60000, "Chandrapur")
    e2 = Emp ("Neha", 50000, "Aurangabad")
    e3 = Emp ("Milind", 70000, "Banglore")
    e1.Display()
    e2.Display()
    
    try:
        e1.EmpSal = -1000
    except InvalidSalException as salex:
        print(salex.message)
    e1.Display()
    
    print(Emp.GetEmpCount())
#     del e1
    
    oel = OnsiteEmp("sid",60000,"hydrabad","v123")
    oel.Display()
    
    print("count:", OnsiteEmp.GetEmpCount())
    
    print("is e1 > e2", e1>e2)
    
    e4 = Emp ("sonu", 35000, "Banglore")
    e5 = Emp ("monu", 75000, "Banglore")
    oe2 = OnsiteEmp("monu", 75000, "Banglore","v101")
    
    elist=[e1,oel,e2,e4,e5,oe2]
    
    print(elist)
    
    elist.sort(key=empsortattr)
    print(elist)
    
    print("print e1",e1)
```
